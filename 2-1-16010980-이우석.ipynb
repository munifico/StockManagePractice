```python
import finterstellar as fs
```

```python
df = fs.get_price('AAPL', start_date='2010-01-01', end_date='2022-3-7')
```

```python
df
```

```python
fs.draw_chart(df, right='AAPL')
```

```python
fs.rsi(df, w=14)
fs.draw_chart(df, left='rsi', right='AAPL')
```

```python
fs.indicator_to_signal(df, factor='rsi', buy=40, sell=60)
```

```python
fs.position(df)
fs.draw_chart(df, left='rsi', right='position_chart')
```

```python
df1 = fs.get_price('AAPL', start_date='2020-01-01', end_date='2020-12-31')
```

```python
fs.rsi(df1, w=14)
fs.draw_chart(df1, left='rsi', right='AAPL')
```

```python
fs.indicator_to_signal(df1, factor='rsi', buy=40, sell=60)
```

```python
fs.position(df1)
fs.draw_chart(df1, left='rsi', right='position_chart')
```

```python
fs.evaluate(df1, cost=0.001)
fs.draw_chart(df1, left='acc_rtn_dp', right='AAPL')
```
